# Data loading

## Create DataFrame from CSV file

* To avoid using memory to guest data type, it is recommended to pass the type : https://stackoverflow.com/questions/24251219/pandas-read-csv-low-memory-and-dtype-options

```python
import pandas as pd

#load dataframe from csv
df = pd.read_csv("/data/time_series_covid19_deaths_global_narrow.csv", dtype={
    "Province/State": "str", 
    "Country/Region": "str"
})

df.info()
```

## Understand distribution
### Parent child columns

```python
columns = ['Country/Region', 'Province/State', 'Lat', 'Long']

# List all unique tuple from many columns & sort
error_by_service = df[columns].drop_duplicates().sort_values(by = columns).reset_index(drop=True)
error_by_service
```

```python
# aggregate on 'error_code' with method 'count'
df.groupby(
    ['Country/Region', 'Province/State']).agg({
        'Value': {
            'occurence' : 'sum'
        }
    }
)
```
